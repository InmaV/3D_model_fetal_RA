```python
import numpy as np
import pyvista as pv
from scipy.spatial import cKDTree

## Fixed wall: STL/VTK right trium points ##
input_RA = "./Meshes/RA/reference_RA.stl"
input_total = "./Meshes/Total/reference.stl"
input_EV = "./Meshes/EV/reference_EV.stl"
output_vtk = "reference.vtk"

y_tolerance = 0.35
distance_threshold = 2.2

# --- Load RA mesh and find points near the minimum y (tricuspid valve)
mesh_RA = pv.read(input_RA)
points_RA = mesh_RA.points
y_coords_RA = points_RA[:, 1]
y_min = np.min(y_coords_RA)
points_TV = points_RA[np.abs(y_coords_RA - y_min) <= y_tolerance]

# --- Find points opposite to the tricuspid valve
y_mean = 0.9*(np.max(y_coords_RA)-np.min(y_coords_RA))+np.min(y_coords_RA) 
points_wall = points_RA[y_coords_RA >= y_mean]

# --- Load EV mesh 
mesh_EV = pv.read(input_EV)
points_EV = mesh_EV.points

# --- Build a KDTree for efficient nearest-neighbor search
tree_RA = cKDTree(points_RA)
tree_TV = cKDTree(points_TV)
tree_wall = cKDTree(points_wall)
tree_EV = cKDTree(points_EV)

# --- Load total mesh
mesh_total = pv.read(input_total)
points_total = mesh_total.points

# --- Find points of total mesh that do not correspond to RA
indices = tree_RA.query_ball_point(points_total, r = distance_threshold)
RA = np.array([len(i) > 0 for i in indices]) 
no_RA = np.array([len(i) == 0 for i in indices]) 


# --- Find points of total mesh that correspond to TV
indices = tree_TV.query_ball_point(points_total, r = distance_threshold)
TV = np.array([len(i) > 0 for i in indices])  

# --- Find points of total mesh that correspond to the wall
indices = tree_wall.query_ball_point(points_total, r = distance_threshold)
wall = np.array([len(i) > 0 for i in indices])  

# --- Find points of total mesh that correspond to EV
indices = tree_EV.query_ball_point(points_total, r = distance_threshold)
EV = np.array([len(i) > 0 for i in indices])  
no_RA_no_EV = no_RA & ~EV
IVC = no_RA_no_EV & (points_total[:, 2] < 80)
SVC = no_RA_no_EV & (points_total[:, 2] > 180)
FO = no_RA_no_EV & ~IVC & ~SVC

# --- Add scalar field and export full mesh with faces ---
mesh_total["RA"] = (RA).astype(np.uint8)
mesh_total["IVC"] = (IVC).astype(np.uint8)
mesh_total["SVC"] = (SVC).astype(np.uint8)
mesh_total["FO"] = (FO).astype(np.uint8)
mesh_total["eustachian_valve"] = (EV).astype(np.uint8)
mesh_total["wall"] = (wall).astype(np.uint8)
mesh_total["tricuspid_valve"] = (TV).astype(np.uint8)
mesh_total["no_RA_no_EV"] = (no_RA_no_EV).astype(np.uint8)
mesh_total.points /= 10  # change units to mm 
mesh_total.save(output_vtk)
```

```python
z_coords = points_total[no_RA_no_EV, 2]

plt.hist(z_coords,bins=500)
plt.hist(np.ones(550)*8,bins=10)
plt.hist(np.ones(550)*18,bins=10)
plt.show() 

print(f"{sum(IVC)} + {sum(SVC)} + {sum(FO)} = {sum(no_RA)}")
```
